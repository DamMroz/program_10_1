# Import modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

# Input data

```python
dataset = pd.read_csv('titanic.csv')
```

# Calculate mean, median and mode

```python
fare_values=dataset['Fare']
mean_fare = np.round(np.mean(fare_values))
median_fare = np.round(np.median(fare_values))
mode_fare = stats.mode(fare_values)
```

```python
mean_fare, median_fare, mode_fare
```

#### The mean values is 32, the median is 14 and the mode amounts to 8.05.

# Calculate quartiles

```python
q0 = np.round(np.quantile(fare_values, 0.0))
q1 = np.round(np.quantile(fare_values, 0.25))
q3 = np.round(np.quantile(fare_values, 0.75))
q4 = np.round(np.quantile(fare_values, 1.0))
```

```python
q0,q1,q3,q4
```

#### The mininum is 0, the first quartile q1 is 8, the third quartile q3 amounts to 31 and the maximum is 512.

# Plot the corresponding boxplot

```python
fig = plt.figure()
fig.suptitle('Fare distribution', fontsize=14, fontweight='bold')
ax = fig.add_subplot(111)
ax.boxplot(fare_values)
plt.xticks([1], ['Fare_data'])
ax.set_ylabel('Fare')
plt.show()
```

# Plot the corresponding histogram

```python
plt.hist(fare_values, bins=20)
plt.axvline(x = mean_fare, color='red', label = 'mean')
plt.axvline(x = median_fare, color='green', label = 'median')
plt.axvline(x = q1, color='black', label = 'q1')
plt.axvline(x = q3, color='blue', label = 'q3')
plt.legend(loc='upper right')
plt.xlabel('Fare')
plt.ylabel('Amount')
plt.show()
```
